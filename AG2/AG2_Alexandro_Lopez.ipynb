```
#Decorador para calcular_tiempo a la función para la cual queremos calcular el tiempo de ejecución(Isabel Vázquez)

from time import time
#Función para calcular el tiempo de ejecución
def calcular_tiempo(f):
   
    def wrapper(*args, **kwargs):        
        inicio = time()       
        resultado = f(*args, **kwargs)       
        tiempo = time() - inicio
        print("Tiempo de ejecución para algoritmo: "+str(tiempo))
        return resultado
    
    return wrapper
from time import time
```

## Dos puntos más cercanos

```
import random
import math
```

```
# se define la función que dará una distancia unidimensional
def dist(A,B):
  if type(A) is int or type(A) is float:
    return abs(B-A)
  else:
    return math.sqrt(sum([(A[i]-B[i])**2 for i in range(len(A))   ])) # distancia euclidiana
  
dist(3,5) # ejemplo de distancia unidimensional
dist((1,3),(2,5)) # ejemplo de distancia bidimensional
  
```

```
# se genera una lista de pares de puntos aleatorios con un número máximo de N elementos
N=4000
LISTA_2D= [  (random.randrange(1,N*10),random.randrange(1,N*10)) for _ in range(N)]
print(LISTA_2D)
```

## Distancia por el método de fuerza bruta

```
# @calcular_tiempo # cálculo de tiempo con el decorador
def dist_fb(L):
  mejor_dist = 100000e10  # se inicializa con un valor alto para garantizar que cualquier par de puntos lo pueden cumplir
  A,B=(),() # valores vacíos donde se almacenarán los mejores elementos encontrados
  for i in range(len(L)):
    for j in range(i+1,len(L)):
      if dist(L[i],L[j])< mejor_dist: # si se encuentra una mejor distancia, se asigna a los valores A y B
        A,B=L[i],L[j]
        mejor_dist=dist(L[i],L[j])
  return[A,B]

dist_fb(LISTA_2D)
```

## Divide y vencerás

```
def dist_dv(L):
  if len(L)<10: # si son menos de 10 elementos, usamos fuerza bruta
    return dist_fb(L)
  
  lista_izq = sorted(L,key=lambda x:x[0])[:len(L)//2] # si son más de 10 elementos se usa divide y vencerás
  lista_der = sorted(L,key=lambda x:x[0])[len(L)//2:]
  
  puntos_lista_izq=dist_dv(lista_izq) # se usa de forma recursiva la función divide y vencerás
  puntos_lista_der=dist_dv(lista_der)
  
  return dist_fb(puntos_lista_izq + puntos_lista_der)

@calcular_tiempo
def LANZA(L):
  return dist_dv(L)

sol=LANZA(LISTA_2D[:100])

print(sol)
```

## Precios en el problema de paseo por el río

```
TARIFAS = [
[0,5,4,3,999,999,999],
[999,0,999,2,3,999,11],
[999,999, 0,1,999,4,10],
[999,999,999, 0,5,6,9],
[999,999, 999,999,0,999,4],
[999,999, 999,999,999,0,3],
[999,999,999,999,999,999,0]
]

def Precios(tarifas):
  N= len(tarifas[0])
  
  precios=[ [9999]*N for i in [9999]*N ]
  ruta = [ [""]*N for i in [9999]*N]
  
  for i in range(N-1):
    for j in range(i+1,N):
      MIN = tarifas[i][j]
      ruta[i][j]=i
      
      for k in range(i,j):
        if precios[i][k]+ tarifas[k][j]< MIN:
          MIN = min (MIN,precios[i][k]+tarifas[k][j])
          ruta[i][j]=k
      precios[i][j]=MIN
  return precios,ruta

CosteFin, RutaFin = Precios(TARIFAS)
print(CosteFin)
print(RutaFin)

def calcular_ruta(RUTA, desde, hasta): # crea la recursividad y  despliega la ruta final
  if desde == hasta:
    #print("Ir a :" + str(desde))
    return desde 
  else:
    return str(calcular_ruta(RUTA, desde, RUTA[desde][hasta])) +  ',' + str(RUTA[desde][hasta]) 

print("\nLa ruta es:")  
calcular_ruta(RutaFin, 0,6)
```
